```python
import pandas as pd
from src.modeling import train_model, save_model
```

# 📌 Load Processed Dataset

```python
df = pd.read_csv('../data/processed/daily_sales.csv')
df.head()
```

# 📌 Sort Data by Date (Important for Time Series)

```python
df = df.sort_values('Date')
df.reset_index(drop=True, inplace=True)

df.head()
```

# 📌 Define Features and  Target Variable

```python
features = [
    'Quantity',
    'Day',
    'Month',
    'Weekday',
    'Weekend',
    'Lag_1',
    'Lag_2',
    'Lag_3',
    'Rolling_Mean_3',
    'Rolling_Mean_7'
]

X = df[features]
y = df['Total_Revenue']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

# 📌Time-Based Set

```python
split = int(len(df) * 0.8)

X_train = X[:split]
y_train = y[:split]
X_test = X[split:]
y_test = y[split:]

print("Test size:", X_test.shape)
print("Train size:", X_train.shape)
```

# 📌 Train Model

```python
model = train_model(X_train, y_train)

print("Model trained successfully")
```

# 📌 Save Trained Model

```python
save_model(model, '../models/final_model.pkl')

print("Model saved successfully in models/final_model.pkl")
```

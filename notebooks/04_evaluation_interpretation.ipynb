# 📌 Import Libraries & Evaluation Functions

```python
import sys
import os
sys.path.append(os.path.abspath(".."))
```

```python
import pandas as pd
from src.modeling import load_model
from src.evaluation import evaluate, plot_actual_vs_predicted, plot_residuals
import matplotlib.pyplot as plt
import seaborn as sns
```

# 📌 Load Processed Dataset

```python
df = pd.read_csv('../data/processed/daily_sales.csv')
```

# 📌 Sort Data by Date (Important for Time Series Evaluation)

```python
df = df.sort_values('Date')
df.reset_index(drop=True, inplace=True)

df.head()
```

# 📌 Define Features & Target Variable

```python
features = [
    'Quantity',
    'Day',
    'Month',
    'Weekday',
    'Weekend',
    'Lag_1',
    'Lag_2',
    'Lag_3',
    'Rolling_Mean_3',
    'Rolling_Mean_7'
]

X = df[features]
y = df['Total_Revenue']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)
```

# 📌 Create Time-Based Set

```python
split = int(len(df) * 0.8)

X_train = X[:split]
y_train = y[:split]
X_test = X[split:]
y_test = y[split:]

print("Test size:", X_test.shape)
print("Train size:", X_train.shape)
```

# 📌 Load Pre-trained Model

```python
model = load_model('../models/final_model.pkl')

print("Model loaded successfully")
```

# 📌 Evaluate Model Performance

```python
preds = evaluate(model, X_test, y_test)
```

# 📌 Plot Actual vs Predictted

```python
plot_actual_vs_predicted(y_test, preds)
plt.savefig("../images/Actual VS Predicted Sales Revenue.png", dpi=300)
plt.show()
plt.close()
```

# 📌 Plot Residual Distribution

```python
plot_residuals(y_test, preds)
```
